```python
def create_array(self):
    print("\n1. Create 1D Array")
    print("2. Create 2D Array")
    print("3. Create 3D Array")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        size = int(input("Enter size of array: "))
        elements = list(map(int, input("Enter elements: ").split()))

        self.arr = np.array(elements)

    elif choice == 2:
        rows = int(input("Enter rows: "))
        cols = int(input("Enter columns: "))

        total = rows * cols

        elements = list(map(int, input(f"Enter {total} elements: ").split()))

        self.arr = np.array(elements).reshape(rows, cols)
        
    elif choice == 3:
        x = int(input("Enter blocks: "))
        y = int(input("Enter rows: "))
        z = int(input("Enter columns: "))

            
        total = x * y * z

        elements = list(map(int, input(f"Enter {total} elements: ").split()))

        self.arr = np.array(elements).reshape(x, y, z)

    else:
        print("Invalid Choice")

    print("\nArray Created Successfully")
    print(self.arr)
```

```python
def indexing_slicing(self):

        if self.arr is None:
            print("Please create array first")
            return

        print("\n1. Indexing")
        print("2. Slicing")

        choice = int(input("Enter your choice: "))

        if choice == 1:

            if self.arr.ndim == 1:
                index = int(input("Enter index: "))
                print("Element:", self.arr[index])

            elif self.arr.ndim == 2:

                row = int(input("Enter row index: "))
                col = int(input("Enter column index: "))

                print("Element:", self.arr[row][col])

        elif choice == 2:

            if self.arr.ndim == 1:

                start = int(input("Enter start index: "))
                end = int(input("Enter end index: "))
                
                print(self.arr[start:end])

            elif self.arr.ndim == 2:

                r1 = int(input("Enter row start: "))
                r2 = int(input("Enter row end: "))
                c1 = int(input("Enter column start: "))
                c2 = int(input("Enter column end: "))

                print(self.arr[r1:r2 , c1:c2])
```

```python
def mathematical_operations(self):

        if self.arr is None:
            print("Please create array first")
            return

        print("\n1. Addition")
        print("2. Subtraction")
        print("3. Multiplication")
        print("4. Division")

        choice = int(input("Enter your choice: "))
        total = self.arr.size

        elements = list(map(int, input(f"Enter {total} elements for second array: ").split()))

        arr2 = np.array(elements).reshape(self.arr.shape)

        if choice == 1:
            print("\nAddition Result")
            print(self.arr + arr2)

        elif choice == 2:
            print("\nSubtraction Result")
            print(self.arr - arr2)

        elif choice == 3:
            print("\nMultiplication Result")
            print(self.arr * arr2)
            
        elif choice == 4:
            print("\nDivision Result")
            print(self.arr / arr2)
```

```python
def combine_split(self):

        if self.arr is None:
            print("Please create array first")
            return

        print("\n1. Combine Arrays")
        print("2. Split Arrays")

        choice = int(input("Enter your choice: "))

        if choice == 1:

            total = self.arr.size

            elements = list(map(int, input(f"Enter {total} elements for second array: ").split()))
            arr2 = np.array(elements).reshape(self.arr.shape)

            result = np.vstack((self.arr, arr2))

            print("\nCombined Array")
            print(result)

        elif choice == 2:

            parts = int(input("Enter number of parts: "))

            result = np.array_split(self.arr, parts)

            print("\nSplit Arrays")

            for i in result:
                print(i)
```

```python
def search_sort_filter(self):

    if self.arr is None:
        print("Please create array first")
        return
    print("\n1. Search")
    print("2. Sort")
    print("3. Filter")

    choice = int(input("Enter your choice: "))
    if choice == 1:
        value = int(input("Enter value to search: "))
        print(np.where(self.arr == value))

    elif choice == 2:

        print("\nSorted Array")
        print(np.sort(self.arr))

    elif choice == 3:

        value = int(input("Show values greater than: "))
        print(self.arr[self.arr > value])
```

```python
def aggregates_statistics(self):
    print("Aggregate Statistics")

    if self.arr is None:
        print("Please create array first")
        return

    print("\n1. Sum")
    print("2. Mean")
    print("3. Median")
    print("4. Standard Deviation")
    print("5. Variance")
    print("6. Minimum")
    print("7. Maximum")
    print("8. Percentile")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        print("Sum =", np.sum(self.arr))

    elif choice == 2:
        print("Mean =", np.mean(self.arr))

    elif choice == 3:
        print("Median =", np.median(self.arr))

    elif choice == 4:
        print("Standard Deviation =", np.std(self.arr))

    elif choice == 5:
        print("Variance =", np.var(self.arr))

    elif choice == 6:
           print("Minimum =", np.min(self.arr))

    elif choice == 7:
        print("Maximum =", np.max(self.arr))
        
    elif choice == 8:
        p = int(input("Enter percentile value: "))
        print(np.percentile(self.arr, p))
```

```python
def correlation(self):

    if self.arr is None:
        print("Please create array first")
        return

    total = self.arr.size

    elements = list(map(int, input(f"Enter {total} elements for second array: ").split()))

    arr2 = np.array(elements)
    arr1 = self.arr.flatten()
    
    print("\nCorrelation Matrix")
    print(np.corrcoef(arr1, arr2))
```
